# 2. 작업형 제2유형 : 모형 구축 및 평가 영역
- 고객 3,500명에 대한 학습용 데이터(y_train.csv, X_train.csv)를 이용하여 성별예측 모형을 만든 후, 이를 평가용 데이터(X_test.csv)에 적용하여 얻은 2,482명 고객의 성별 예측값(남자일 확률 = gender:1)을 다음과 같은 형식의 CSV 파일로 생성하시오.(제출한 모델의 성능은 ROC-AUC 평가지표에 따라 채점)



```python
import pandas as pd 
X_train = pd.read_csv('X_train.csv',encoding='cp949')
y_train = pd.read_csv('y_train.csv',encoding='cp949')

X_test = pd.read_csv('X_test.csv',encoding='cp949')
```

```python
X_train['환불금액'].fillna(0, inplace=True)
```

```python
X_train.info()
```

```python
X_test['환불금액'].fillna(0, inplace=True)
```

```python
X_test.info()
```

```python
X_train = pd.get_dummies(X_train, columns=['주구매상품', '주구매지점'])
```

```python
X_train.set_index('cust_id', inplace=True)
```

```python
X_train
```

```python
X_test = pd.get_dummies(X_test, columns=['주구매상품', '주구매지점'])
```

```python
X_test.set_index('cust_id',inplace=True)
```

```python
X_test
```

```python
y_train.set_index('cust_id',inplace=True)
```

```python
y_train
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier() 
```

```python
clf.fit(X_train, y_train)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train.columns[X_train.columns.isin(X_test)==False]
```

```python
X_test['주구매상품_소형가전'] = 0 
```

```python
X_test.shape
```

```python
result = clf.predict_proba(X_test)[:,1] 
```

```python
pd.DataFrame(result).to_csv("result.csv", index=False)
```
